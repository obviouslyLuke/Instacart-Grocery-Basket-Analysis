## 5. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.
## If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
## If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
## If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


```python
#Creating column 'user_order_behaviour'
ords_prods_merge['user_order_behaviour'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.max)
```

```python
# Deriving columns with loc(). This will assign labels depending on the count in the in the 'user_order_behaviour' column
ords_prods_merge.loc[ords_prods_merge['user_order_behaviour'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['user_order_behaviour'] > 10) & (ords_prods_merge['user_order_behaviour'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['user_order_behaviour'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

### 6.The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
### If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
ords_prods_merge['user_spending_behaviour'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['user_spending_behaviour'] < 10, 'spending_flag'] = 'High Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['user_spending_behaviour'] >= 10, 'spending_flag'] = 'Low Spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```
