```python
# Import dictionaries
import pandas as pd
import numpy as np
import os
```

## 1. Import the Data

```python
## defining directory path
path = r'C:\Users\luke_\Desktop\CF Datacourse\Python_exercise4\09-2024 Instacart Basket Analysis\02_data'
```

```python
#Importing products_checked.csv as df_prods
df_prods = pd.read_csv(os.path.join(path,'prepared_data', 'products_checked.csv'), index_col = False) 
```

```python
#Importing orders_products_combined.pkl as df_ords_prods_combined
df_ords_prods_combined = pd.read_pickle(os.path.join(path, 'prepared_data', 'orders_products_combined.pkl'))
```

## 3.Merge df_prods with df_ords_prods_combined.

```python
# There are columns which we can drop in df_prods.
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
# Likewise in df_ords_prods_combined there are columns we can drop
df_ords_prods_combined = df_ords_prods_combined.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0', '_merge'])
```

```python
# Merge df_prods and df_ords_prods_combined
df_merged = df_prods.merge(df_ords_prods_combined, on = 'product_id', indicator = True)
```

## 5.Export the file.

```python
df_merged.to_pickle(os.path.join(path, 'prepared_data', 'ords_prods_merged.pkl'))
```
